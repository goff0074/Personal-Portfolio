```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
alpha = np.linspace(0, 1, 100)

x1 = 0 + alpha * 1
x2 = 2 + alpha * (-1)

f_x = x1**3 - 5 * x1 * x2 + 6 * x2**2
f_x_first = 24 - 34 * alpha
f_x_second = 22 - 34 * alpha - 11 * alpha ** 2
```

```python
plot.figure(figsize = (10, 10))
plot.plot(alpha, f_x, label = r'f(x)', color = 'red')
plot.plot(alpha, f_x_first , label = r'First Order Taylor Approximation', color = 'blue')
plot.plot(alpha, f_x_second, label = r'Second Order Taylor Approximation', color = 'green')
plot.title('Graph of f(x) and its First and Second Order Taylor Approximation') 
plot.xlabel('X')
plot.ylabel('Y')
```
